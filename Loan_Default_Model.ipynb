# **Calculating Cramer's V matrix for the correlation between categorical columns:**

```python
from scipy.stats import chi2_contingency
# Function to check for zero variance in categorical variable
def has_zero_variance(series):
    return len(series.unique()) == 1

# Function to calculate Cramer's V with handling for zero variance=unique
def cramers_v(x, y):
    if has_zero_variance(x) or has_zero_variance(y):
        return np.nan  # Return NaN if either variable has zero variance
    else:
        confusion_matrix = pd.crosstab(x, y)
        chi2, _, _, _ = chi2_contingency(confusion_matrix)
        n = confusion_matrix.sum().sum()
        phi2 = chi2 / n
        r, k = confusion_matrix.shape
        phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
        rcorr = r - ((r - 1) ** 2) / (n - 1)
        kcorr = k - ((k - 1) ** 2) / (n - 1)
        return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Calculate Cramer's V matrix for all pairs of categorical variables
cat_cols = df.select_dtypes(include=['object']).columns
cramers_v_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols, dtype=float)
for i in range(len(cat_cols)):
    for j in range(i, len(cat_cols)):
        if i != j:
            v = cramers_v(df[cat_cols[i]], df[cat_cols[j]])
            cramers_v_matrix.at[cat_cols[i], cat_cols[j]] = v
            cramers_v_matrix.at[cat_cols[j], cat_cols[i]] = v

# Display Cramer's V matrix
print("Cramer's V Matrix:")
print(cramers_v_matrix)
```

```python
# Identify highly associated columns based on a threshold (e.g., 0.5)
threshold = 0.5
highly_associated_cols = set()
for i in range(len(cat_cols)):
    for j in range(i + 1, len(cat_cols)):
        if cramers_v_matrix.iloc[i, j] >= threshold:
            highly_associated_cols.add(cat_cols[j])

# Drop highly associated columns from the DataFrame
df.drop(columns=highly_associated_cols, inplace=True)

print("Remaining cat columns after dropping highly associated columns:")
print(df.select_dtypes(include=['object']).columns.tolist())
```

# **categorize employment titles for one-hot encoding based on income levels**

# Reducing the unique sub-categories in 'emp_title' column:

```python
# Define thresholds based on overall median income
overall_median_income = df['annual_inc'].median()
low_income_threshold = overall_median_income / 2
high_income_threshold = overall_median_income * 1.5


# Create a function to categorize jobs based on income
def categorize_job(title):
    annual_income = df[df['emp_title'] == title]['annual_inc'].mean()# mean of annual_inc after filtering according to the given job title
    if annual_income < low_income_threshold:
        return 'low_income_job'
    elif annual_income < high_income_threshold:
        return 'moderate_income_job'
    else:
        return 'high_income_job'

# Apply the categorization to the 'emp_title' column
df['emp_title'] = df['emp_title'].apply(categorize_job)
```

# **Converting the date-related columns into age**

```python
from datetime import datetime

# Convert date columns to datetime format
date_columns = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='%b-%Y')

# Calculate current date
current_date = datetime.now()

# Calculate age in years
for col in date_columns:
    df[col] = (current_date - df[col]).dt.days / 365.25
```

# **Evaluation Metrics:**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Function to evaluate model performance
def evaluate_model(name, y_test, y_pred):
    print(f"Evaluation for {name}:")

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Calculate precision, recall, and f1-score
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Print results
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")

    # Print confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(cm)

    # Print classification report
    report = classification_report(y_test, y_pred, zero_division=0)
    print("Classification Report:")
    print(report)
    print("-------------------------------------")

# Example usage with your models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    evaluate_model(name, y_test, y_pred)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from imblearn.under_sampling import RandomUnderSampler

# Apply random undersampling to the training data
undersampler = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train, y_train)

# Initialize the XGBoost model
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', n_estimators=100, max_depth=3, learning_rate=0.1)

# Define scoring metrics for cross-validation
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='weighted'),
    'recall': make_scorer(recall_score, average='weighted'),
    'f1': make_scorer(f1_score, average='weighted')
}

# Perform cross-validation with the resampled data
cv_results = cross_validate(xgb_model, X_train_resampled, y_train_resampled, cv=5, scoring=scoring)

# Print the cross-validation results
print("Cross-validation results:")
print(f"Accuracy: {cv_results['test_accuracy'].mean():.4f}  {cv_results['test_accuracy'].std():.4f}")
print(f"Precision: {cv_results['test_precision'].mean():.4f}  {cv_results['test_precision'].std():.4f}")
print(f"Recall: {cv_results['test_recall'].mean():.4f}  {cv_results['test_recall'].std():.4f}")
print(f"F1 Score: {cv_results['test_f1'].mean():.4f}  {cv_results['test_f1'].std():.4f}")
```

# **Learning Curve before the Hyperparameter Tuning**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.model_selection import train_test_split

# Define the training sizes to use in the learning curve
train_sizes = np.linspace(0.1, 1.0, 10)  # Use 10 evenly spaced training set sizes from 10% to 100%

# Compute learning curves
train_sizes, train_scores, val_scores = learning_curve(xgb_model, X_train_resampled, y_train_resampled, cv=5, train_sizes=train_sizes, scoring='accuracy', n_jobs=-1)

# Calculate mean and standard deviation of training scores and validation scores across folds
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)
val_scores_std = np.std(val_scores, axis=1)

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, val_scores_mean, 'o-', color="g", label="Cross-validation score")
plt.xlabel("Training examples")
plt.ylabel("Score")
plt.title("Learning Curve (XG Boost Model)")
plt.legend(loc="best")
plt.show()
```

# **Hyperparameter Tuning**

```python
from sklearn.model_selection import cross_validate, RandomizedSearchCV, train_test_split
from scipy.stats import uniform, randint
from collections import Counter
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.under_sampling import RandomUnderSampler
from xgboost import XGBClassifier
```

```python
# Define hyperparameter distributions for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(50, 200),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'min_child_weight': randint(1, 6),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4)
}

# Define scoring metrics for cross-validation
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='weighted'),
    'recall': make_scorer(recall_score, average='weighted'),
    'f1': make_scorer(f1_score, average='weighted')
}

# Perform Randomized Search with cross-validation
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    scoring=scoring,
    n_jobs=-1,
    verbose=2,
    random_state=42,
    refit='f1'  # Specify the refit metric
)
random_search.fit(X_train, y_train)

# Get the best model
best_xgb_model = random_search.best_estimator_
print(f"Best XGBoost model: {random_search.best_params_}")

# Evaluate the best model with cross-validation
cv_results_best = cross_validate(best_xgb_model, X_train_resampled, y_train_resampled, cv=5, scoring=scoring)

# Print the cross-validation results for the best model
print("Cross-validation results for the best model:")
print(f"Accuracy: {cv_results_best['test_accuracy'].mean():.4f}  {cv_results_best['test_accuracy'].std():.4f}")
print(f"Precision: {cv_results_best['test_precision'].mean():.4f}  {cv_results_best['test_precision'].std():.4f}")
print(f"Recall: {cv_results_best['test_recall'].mean():.4f}  {cv_results_best['test_recall'].std():.4f}")
print(f"F1 Score: {cv_results_best['test_f1'].mean():.4f}  {cv_results_best['test_f1'].std():.4f}")
```
